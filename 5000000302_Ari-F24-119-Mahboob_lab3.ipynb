## Task 3

```python
lib_books = {
    "Atomic Habits": ("James Clear", 2018, "Self-help"),
    "The Psychology of Money": ("Morgan Housel", 2020, "Finance"),
    "Rich Dad Poor Dad": ("Robert Kiyosaki", 1997, "Personal Finance"),
    "Ikigai": ("Héctor García & Francesc Miralles", 2016, "Self-help")
}

def add_book(library, title, author, year, genre):
    library[title] = (author, year, genre)
    print(f"Book '{title}' added successfully!")

def search_book(library, title):
    if title in library:
        author, year, genre = library[title]
        print(f"\nTitle: {title}\nAuthor: {author}\nYear: {year}\nGenre: {genre}")
    else:
        print(f"Book '{title}' not found in library.")

def delete_book(library, title):
    if title in library:
        library.pop(title)
        print(f"Book '{title}' deleted successfully.")
    else:
        print(f"Book '{title}' not found, so cannot delete.")

while True:
    print("\n--- Library Menu ---")
    print("1. Add a book")
    print("2. Search for a book")
    print("3. Delete a book")
    print("4. View all books")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ")

    if choice == "1":
        title = input("Enter book title: ")
        author = input("Enter author: ")
        year = int(input("Enter publication year: "))
        genre = input("Enter genre: ")
        add_book(lib_books, title, author, year, genre)

    elif choice == "2":
        title = input("Enter book title to search: ")
        search_book(lib_books, title)

    elif choice == "3":
        title = input("Enter book title to delete: ")
        delete_book(lib_books, title)

    elif choice == "4":
        print("\n--- All Books in Library ---")
        for t, (a, y, g) in lib_books.items():
            print(f"{t} → {a}, {y}, {g}")

    elif choice == "5":
        print("Exiting Library System. Goodbye!")
        break

    else:
        print("Invalid choice. Please try again.")
```
