```python
#Step 1: Import the Needed libraries
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python

#Step 2: Make a Small Dataset

 # Study hours (input) and exam scores (output)
study_hours = np.array([1, 2, 3, 4, 5, 6, 7, 8])
exam_scores = np.array([35, 40, 50, 55, 65, 70, 80, 90])
# Quick check
print("Study Hours:", study_hours)
print("Exam Scores:", exam_scores)
```

```python
 # Build a simple ANN
 model = Sequential()
 # Input + Hidden layer (64 neurons). input_dim=1 → one input feature: Study Hours
 model.add(Dense(64, input_dim=1, activation='relu'))
 # Output layer (1 neuron) → predicts the exam score
 model.add(Dense(1, activation='linear'))
 # Compile: tell the model how to learn
 # optimizer='adam' is like a helpful teacher
# loss='mean_squared_error' measures how far off predictions are
 model.compile(optimizer='adam', loss='mean_squared_error')
 # Show the model summary
 model.summary()
```

```python
# Step 4: Train the Model (Let It Learn!) 
history = model.fit(study_hours, exam_scores, epochs=800, verbose=1)
# Plot training loss to see learning progress
plt.plot(history.history['loss'])
plt.title('Training Loss Over Time')
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE)')
plt.show()
```

```python
new_study_hours=np.array([2,4,6,8])
pre= model.predict(new_study_hours)

print("Marks", pre.flatten().round(1))
```

```python
#Step 6
# Visualize What the Model Learned
# Plot actual data points
plt.scatter(study_hours, exam_scores, color='blue', label='Actual Data')
# Plot model predictions as a red line
line_hours = np.linspace(1, 8, 50).astype(float)
line_preds = model.predict(line_hours)
plt.plot(line_hours, line_preds, color='red', label='Predicted Line')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.title('ANN Prediction: Study Hours → Exam Score')
```
