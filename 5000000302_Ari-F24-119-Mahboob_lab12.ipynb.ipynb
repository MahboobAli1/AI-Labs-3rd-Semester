```python
# Step 1 – Create a Small Text Dataset
texts = [
 "I love this movie!",
 "This product is terrible",
 "Amazing experience",
 "I hate this!",
 "Very good service",
 "Worst food ever"
 ]
 # 1 = Positive, 0 = Negative
labels = [1, 0, 1, 0, 1, 0]

#Step 2 – Clean the Text
import re
def clean_text(text):
 # 1) convert to lowercase
    text = text.lower()
 # 2) keep only letters and spaces
    text = re.sub(r"[^a-zA-Z ]", "", text)
    return text
cleaned = [clean_text(t) for t in texts]
cleaned
```

```python
#Step 3 – Convert Text to Numbers (Bag of Words)
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(cleaned)
print("Vocabulary (words the model knows):")
print(vectorizer.get_feature_names_out())
print("\nBag of Words representation (as array):")
X.toarray()
 # Create a table for Vocabulary and Bag of Words representation
import pandas as pd
# Convert sparse matrix to dense array
bow_array = X.toarray()
 # Get the vocabulary (words)
words = vectorizer.get_feature_names_out()
 # Create a DataFrame
df_bow = pd.DataFrame(bow_array, columns=words)
 # Display the table
df_bow
```

```python
#Step 4 – Train a Simple Sentiment Classifier
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X, labels)
print("Model trained!")
```

```python
# --- Step 5: Test With Your Own Sentence ---
def predict_sentiment(sentence):
    # 1. Clean the text
    cleaned_sentence = clean_text(sentence)
    
    # 2. Transform it to numbers (using the existing vectorizer)
    vec = vectorizer.transform([cleaned_sentence])
    
    # 3. Predict (0 or 1)
    pred = model.predict(vec)[0]
    
    # 4. Display results
    print("Cleaned sentence:", cleaned_sentence)
    
    if pred == 1:
        print("Prediction: Positive Sentence")
    else:
        print("Prediction: Negative Sentence")

# --- Run the Test ---
user_input = "I love to watch movies!"
predict_sentiment(user_input)
```
