```python
import numpy as np 
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
print ("libraries imported")
```

```python
(train_images, train_labels), (test_images, test_labels) = keras.datasets.mnist.load_data()
```

```python
print("--- Computer View (Pixel Grid) ---")
print(train_images[0])
```

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
print("Normalization complete")
```

```python
model = keras.Sequential()
model.add(keras.layers.Flatten(input_shape=(28, 28)))
model.add(keras.layers.Dense(256, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))
model.compile(optimizer='adam', 
loss='sparse_categorical_crossentropy', 
metrics=['accuracy'])
print("Model Compiled")
```

```python
model.fit(train_images, train_labels, epochs=10)
```

```python
image_id = 1000
picture = test_images[image_id]
real_answer = test_labels[image_id]
plt.imshow(picture, cmap=plt.cm.binary)
plt.title(f"Human wrote: {real_answer}")
plt.show()
prediction = model.predict(np.array([picture]))
predicted_number = np.argmax(prediction[0])
print(f"AI Thinks it is: {predicted_number}")
if predicted_number == real_answer:
 print("CORRECT!")
else:
 print("WRONG!")
```
