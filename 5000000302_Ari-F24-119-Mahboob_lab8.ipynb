```python
def dfs_search(tree, target):
    stack = [tree.root]  # Start from the root node

    while stack:
        node_id = stack.pop()            # Get the node identifier
        node = tree.get_node(node_id)    # Get the actual Node object

        # Step 4: Check if current node matches the target
        if node.tag == target:
            print(f"Node '{target}' found!")
            return True

        # Step 5: Add child nodes to the stack
        for child in tree.children(node.identifier):
            stack.append(child.identifier)

    print(f"Node '{target}' not found.")
    return False

# Step 6: Call the DFS function
dfs_search(tree, 'Ali')
```

```python

 # Import Tree class from treelib
 from treelib import Node, Tree# BFS Implementation using treelib
# -------------------------------

# Step 1: Create a tree
tree = Tree()
tree.create_node("A", "A")  # Root
tree.create_node("B", "B", parent="A")
tree.create_node("C", "C", parent="A")
tree.create_node("D", "D", parent="B")
tree.create_node("E", "E", parent="B")
tree.create_node("F", "F", parent="C")

# Step 2: Define a function to check if queue is empty
def checkqueue(queue):
    if not queue:
        print('Queue is empty')
        return True
    else:
        print('Queue is not empty')
        return False

# Step 3: Define BFS function
def bfs_search(tree, target):
    # Initialize queue with root node
    queue = [tree.get_node("A")]

    # Step 4: Start BFS
    while not checkqueue(queue):   # Check before dequeue
        node = queue.pop(0)        # Dequeue front node
        print(f"Visiting node: {node.tag}")

        # Step 5: Check if current node is target
        if node.tag == target:
            print(f" Node '{target}' found!")
            return True

        # Step 6: Add child nodes to queue
        for child in tree.children(node.identifier):
            queue.append(child)

    # Step 7: If target not found
    print(f" Node '{target}' not found.")
    return False

# Step 8: Call the BFS function
bfs_search(tree, 'C')
```

```python
from treelib import Node, Tree
tree = Tree()
tree.create_node("A", "A")  # Root
tree.create_node("B", "B", parent="A")
tree.create_node("C", "C", parent="A")
tree.create_node("D", "D", parent="B")
tree.create_node("E", "E", parent="B")
tree.create_node("F", "F", parent="C")
def checkqueue(queue):
    if not queue:
        print('Queue is empty')
        return True
    else: 
        print("Queue is not empty ")
        return False

def bfs_search_multiple(tree, targets):
    queue = [tree.get_node("A")]
    parent = {"A": None}  # Keep track of parents for path reconstruction
    found_paths = {}      # Store found targets with their paths

    while not checkqueue(queue):
        node = queue.pop(0)
        print(f"Visiting node: {node.tag}")

        # If node is one of the targets, record its path
        if node.tag in targets:
            path = []
            current = node.identifier
            while current:
                path.insert(0, current)
                current = parent[current]
            found_paths[node.tag] = path
            print(f"Found target '{node.tag}' with path: {' -> '.join(path)}")

        # Enqueue children and record their parent
        for child in tree.children(node.identifier):
            queue.append(child)
            parent[child.identifier] = node.identifier

    # Print summary
    if found_paths:
        print("\n Search completed! Found these targets and paths:")
        for target, path in found_paths.items():
            print(f"{target}: {' -> '.join(path)}")
    else:
        print(" No targets found.")


targets = ['C', 'E', 'F']
bfs_search_multiple(tree, targets)

        
```
