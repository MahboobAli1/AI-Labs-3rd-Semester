## Task 9

```python

arr1 = np.array([[1, 2],
                 [3, 4]])

arr2 = np.array([[5, 6],
                 [7, 8]])

print("Array 1:\n", arr1)
print("Array 2:\n", arr2)

v_merge = np.vstack((arr1, arr2))

h_merge = np.hstack((arr1, arr2))

print("Merged Vertically:\n", v_merge)
print("Merged Horizontally:\n", h_merge)
```
